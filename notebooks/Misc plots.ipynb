```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource, FactorRange
from bokeh.transform import factor_cmap
from bokeh.palettes import Spectral8

output_notebook()
all_metrics = ['Precision (F)', 'Recall (F)', 'F1 (F)', 'Precision (T)', 'Recall (T)', 'F1 (T)', 'Accuracy', 'ROC AUC']
all_data = {'classifiers': classifiers,
        'Precision (F)': [0.87, 0.88, 0.93],
        'Recall (F)': [0.96, 0.96, 0.68],
        'F1 (F)': [0.91, 0.92, 0.79],
        'Precision (T)': [0.70, 0.72, 0.37],
        'Recall (T)': [0.41, 0.46, 0.80],
        'F1 (T)': [0.52, 0.56, 0.51],
        'Accuracy': [0.85, 0.86, 0.71],
        'ROC AUC': [0.89, 0.90, 0.82]}
```

```python
classifiers = ['Decision Tree', 'Random Forest', 'Naive Bayes']
metrics = [
    'Precision (F)', 
    'Recall (F)',
    'F1 (F)',
    'Precision (T)',
    'Recall (T)',
    'F1 (T)',
    'Accuracy',
    'ROC AUC'
]

x = [ (classifier, metric) for classifier in classifiers for metric in metrics ]
counts = sum(zip(
    all_data['Precision (F)'],
    all_data['Recall (F)'],
    all_data['F1 (F)'],
    all_data['Precision (T)'],
    all_data['Recall (T)'],
    all_data['F1 (T)'],
    all_data['Accuracy'],
    all_data['ROC AUC']
        ),())

source = ColumnDataSource(data=dict(x=x, counts=counts))

plot = figure(plot_width=400, plot_height=400, x_range=FactorRange(*x))
plot.vbar(x=x, width=0.5, bottom=0, top=counts, fill_color=factor_cmap('x', palette=Spectral8, factors=metrics, start=1, end=2))
plot.x_range.range_padding = 0.1
plot.xaxis.major_label_orientation = 1

show(plot)
```
