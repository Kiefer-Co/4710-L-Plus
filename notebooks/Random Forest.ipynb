```python
import pandas as pd
dataset = pd.read_csv("pumf-98M0001-E-2016-individuals_F1.csv")
dataset.head()
```

```python
# class label: home language part A - first language write in component
homeLang = dataset.iloc[:,65]
# class label: mother tongue part A - first language write in component
motherTongue = dataset.iloc[:, 96]
# variables
x1, x2, x3 = dataset.iloc[:, 18:65], dataset.iloc[:, 66:96], dataset.iloc[:, 97:-1]
x = pd.concat([x1,x2,x3], axis=1, sort=False)
weights = dataset.iloc[:, 1]
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier = RandomForestClassifier(n_estimators=20, random_state=0)
```

```python
from sklearn.model_selection import train_test_split
weights_train, weights_test, x_train, x_test, homeLang_train, homeLang_test, motherTongue_train, motherTongue_test = train_test_split(weights, x, homeLang, motherTongue)
```

```python
classifier.fit(x_train, homeLang_train, weights_train)
```

```python
homeLang_pred = classifier.predict(x_test)
```

```python
langs=['No non-official language','Aboriginal languages','Italian','Spanish','Portuguese','German','Russian','Polish','Slavic languages','Other European languages','Arabic','Other Afro-Asiatic and African languages','Punjabi (Panjabi)','Urdu','Persian (Farsi)','Other Indo-Iranian languages','Cantonese','Mandarin','Chinese languages','Austro-Asiatic languages','Tagalog (Pilipino, Filipino)','Other East and Southeast Asian languages','Tamil','All other languages','Not available']
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

The Confusion Matrix indicates, given the predicted language (row names), for how many records did the algorith match the actual language. The classification report and accuracy score are other ways to evaluate how well the classification algorithm performed.

```python
confusionMatrixHomeLang = pd.DataFrame(
    confusion_matrix(homeLang_test,homeLang_pred), 
    index=langs, 
    columns=langs
)
```

```python
print(confusionMatrixHomeLang)
print(classification_report(homeLang_test,homeLang_pred,target_names=langs))
print(accuracy_score(homeLang_test, homeLang_pred))
```
